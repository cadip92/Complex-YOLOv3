```python
fig = plt.figure(figsize=(8,5))

ax = Axes3D(fig)

ax.set_xticks(np.arange(0, 65, step=5))
ax.set_yticks(np.arange(60, 100, step=5))

ax.set_xlabel('# Parameters (millions)')
ax.set_ylabel('mAP (%)')

### 3D Scatter plot for models run on RTX 2080Ti
ax.set_zticks(np.arange(14, 32, step=2))

ax.set_zlabel('FPS on Nvidia RTX 2080Ti')

comp_v3 = ax.scatter3D(model_size[1], mAP[1], FPS_slarti[0], c='g', marker='1', s=150)
dep_sep = ax.scatter3D(model_size[2], mAP[2], FPS_slarti[1], c='c', marker='o', s=150)
mobile_v1 = ax.scatter3D(model_size[3], mAP[3], FPS_slarti[2], c='c', marker='x', s=150)
mobile_v12 = ax.scatter3D(model_size[4], mAP[4], FPS_slarti[3], c='m', marker='^', s=150)
mobile_v2 = ax.scatter3D(model_size[5], mAP[5], FPS_slarti[4], c='g', marker='v', s=150)
mobile_v21 = ax.scatter3D(model_size[6], mAP[6], FPS_slarti[5], c='y', marker='<', s=150)
mobile_v22 = ax.scatter3D(model_size[7], mAP[7], FPS_slarti[6], c='c', marker='>', s=150)
mobile_v23 = ax.scatter3D(model_size[8], mAP[8], FPS_slarti[7], c='m', marker='3', s=150)
mobile_v24 = ax.scatter3D(model_size[9], mAP[9], FPS_slarti[8], c='b', marker='*', s=150)
mobile_v25 = ax.scatter3D(model_size[10], mAP[10], FPS_slarti[9], c='k', marker='.', s=150)
mobile_v251 = ax.scatter3D(model_size[11], mAP[11], FPS_slarti[10], c='y', marker='s', s=150)
pelee_v1 = ax.scatter3D(model_size[12], mAP[12], FPS_slarti[11], c='g', marker='d', s=150)
pelee_v22 = ax.scatter3D(model_size[13], mAP[13], FPS_slarti[12], c='b', marker='p', s=150)
pelee_v111 = ax.scatter3D(model_size[14], mAP[14], FPS_slarti[13], c='k', marker='h', s=150)
pelee_v12 = ax.scatter3D(model_size[15], mAP[15], FPS_slarti[14], c='k', marker='4', s=150)
pelee_v121 = ax.scatter3D(model_size[16], mAP[16], FPS_slarti[15], c='c', marker='+', s=150)

# For Pareto Front visualization
inputPoints = [
    [model_size[1], mAP[1], FPS_slarti[0]],
    [model_size[2], mAP[2], FPS_slarti[1]],
    [model_size[3], mAP[3], FPS_slarti[2]],
    [model_size[4], mAP[4], FPS_slarti[3]],
    [model_size[5], mAP[5], FPS_slarti[4]],
    [model_size[6], mAP[6], FPS_slarti[5]],
    [model_size[7], mAP[7], FPS_slarti[6]],
    [model_size[8], mAP[8], FPS_slarti[7]],
    [model_size[9], mAP[9], FPS_slarti[8]],
    [model_size[10], mAP[10], FPS_slarti[9]],
    [model_size[11], mAP[11], FPS_slarti[10]],
    [model_size[12], mAP[12], FPS_slarti[11]],
    [model_size[13], mAP[13], FPS_slarti[12]],
    [model_size[14], mAP[14], FPS_slarti[13]],
    [model_size[15], mAP[15], FPS_slarti[14]],
    [model_size[16], mAP[16], FPS_slarti[15]],
]
###

### 3D Scatter plot for models run on GeForce 940MX
# ax.set_zticks(np.arange(1, 7.2, step=0.5))

# ax.set_zlabel('FPS on Nvidia GeForce 940MX')

# comp_v3 = ax.scatter3D(model_size[1], mAP[1], FPS_laptop[0], c='g', marker='1', s=150)
# dep_sep = ax.scatter3D(model_size[2], mAP[2], FPS_laptop[1], c='c', marker='o', s=150)
# mobile_v1 = ax.scatter3D(model_size[3], mAP[3], FPS_laptop[2], c='c', marker='x', s=150)
# mobile_v12 = ax.scatter3D(model_size[4], mAP[4], FPS_laptop[3], c='m', marker='^', s=150)
# mobile_v2 = ax.scatter3D(model_size[5], mAP[5], FPS_laptop[4], c='g', marker='v', s=150)
# mobile_v21 = ax.scatter3D(model_size[6], mAP[6], FPS_laptop[5], c='y', marker='<', s=150)
# mobile_v22 = ax.scatter3D(model_size[7], mAP[7], FPS_laptop[6], c='c', marker='>', s=150)
# mobile_v23 = ax.scatter3D(model_size[8], mAP[8], FPS_laptop[7], c='m', marker='3', s=150)
# mobile_v24 = ax.scatter3D(model_size[9], mAP[9], FPS_laptop[8], c='b', marker='*', s=150)
# mobile_v25 = ax.scatter3D(model_size[10], mAP[10], FPS_laptop[9], c='k', marker='.', s=150)
# mobile_v251 = ax.scatter3D(model_size[11], mAP[11], FPS_laptop[10], c='y', marker='s', s=150)
# pelee_v1 = ax.scatter3D(model_size[12], mAP[12], FPS_laptop[11], c='g', marker='d', s=150)
# pelee_v22 = ax.scatter3D(model_size[13], mAP[13], FPS_laptop[12], c='b', marker='p', s=150)
# pelee_v111 = ax.scatter3D(model_size[14], mAP[14], FPS_laptop[13], c='k', marker='h', s=150)
# pelee_v12 = ax.scatter3D(model_size[15], mAP[15], FPS_laptop[14], c='k', marker='4', s=150)
# pelee_v121 = ax.scatter3D(model_size[16], mAP[16], FPS_laptop[15], c='c', marker='+', s=150)

# # For Pareto Front visualization
# inputPoints = [
#     [model_size[1], mAP[1], FPS_laptop[0]],
#     [model_size[2], mAP[2], FPS_laptop[1]],
#     [model_size[3], mAP[3], FPS_laptop[2]],
#     [model_size[4], mAP[4], FPS_laptop[3]],
#     [model_size[5], mAP[5], FPS_laptop[4]],
#     [model_size[6], mAP[6], FPS_laptop[5]],
#     [model_size[7], mAP[7], FPS_laptop[6]],
#     [model_size[8], mAP[8], FPS_laptop[7]],
#     [model_size[9], mAP[9], FPS_laptop[8]],
#     [model_size[10], mAP[10], FPS_laptop[9]],
#     [model_size[11], mAP[11], FPS_laptop[10]],
#     [model_size[12], mAP[12], FPS_laptop[11]],
#     [model_size[13], mAP[13], FPS_laptop[12]],
#     [model_size[14], mAP[14], FPS_laptop[13]],
#     [model_size[15], mAP[15], FPS_laptop[14]],
#     [model_size[16], mAP[16], FPS_laptop[15]],
# ]
###

### Calculate Pareto points for the model
def simple_cull(inputPoints, dominates):
    paretoPoints = set()
    candidateRowNr = 0
    dominatedPoints = set()
    while True:
        candidateRow = inputPoints[candidateRowNr]
        inputPoints.remove(candidateRow)
        rowNr = 0
        nonDominated = True
        while len(inputPoints) != 0 and rowNr < len(inputPoints):
            row = inputPoints[rowNr]
            if dominates(candidateRow, row):
                # If it is worse on all features remove the row from the array
                inputPoints.remove(row)
                dominatedPoints.add(tuple(row))
            elif dominates(row, candidateRow):
                nonDominated = False
                dominatedPoints.add(tuple(candidateRow))
                rowNr += 1
            else:
                rowNr += 1

        if nonDominated:
            # add the non-dominated point to the Pareto frontier
            paretoPoints.add(tuple(candidateRow))

        if len(inputPoints) == 0:
            break
    return paretoPoints, dominatedPoints

def dominates(row, candidateRow):
    sum = 0
    if row[0] < candidateRow[0] : sum += 1
    if row[1] >= candidateRow[1] : sum += 1
    if row[2] >= candidateRow[2] : sum += 1
    return sum == len(row)

paretoPoints, dominatedPoints = simple_cull(inputPoints, dominates)

dp = np.array(list(dominatedPoints))
pp = np.array(list(paretoPoints))

pp_plot = ax.scatter(pp[:,0],pp[:,1],pp[:,2], s=160, facecolor=(0,0,0,0), marker='o', edgecolor='r', linewidths=2.3)

plt.legend((comp_v3, dep_sep, mobile_v1, mobile_v12, mobile_v2, mobile_v21, mobile_v22, mobile_v23, mobile_v24, 
            mobile_v25, mobile_v251, pelee_v1, pelee_v11, pelee_v111, pelee_v12, pelee_v121, pp_plot),
           (['Complex_YOLOv3', 
     'Dep_Sep_Complex_YOLO', 
     'Mobile_Complex_YOLOv1', 
     'Mobile_Complex_YOLOv1.2',
     'Mobile_Complex_YOLOv2',
     'Mobile_Complex_YOLOv2.1',
     'Mobile_Complex_YOLOv2.2',
     'Mobile_Complex_YOLOv2.3',
     'Mobile_Complex_YOLOv2.4',
     'Mobile_Complex_YOLOv2.5',
     'Mobile_Complex_YOLOv2.5.1',
     'Pelee_Complex_YOLOv1',
     'Pelee_Complex_YOLOv1.1',
     'Pelee_Complex_YOLOv1.1.1',
     'Pelee_Complex_YOLOv1.2',
     'Pelee_Complex_YOLOv1.2.1',
     'Pareto Point']),
           scatterpoints=1,
           loc='best',
           ncol=1,
           bbox_to_anchor=(1.08, 0.85),
           fontsize=11)

### Pareto plot for eval on GeForce 940MX
# plt.savefig('H:/Thesis_Git/thesis_docs/Final_Thesis/Images/Complex-YOLO/pareto_optimality_laptop.pdf', bbox_inches='tight')

### Pareto plot for eval on RTX 2080 Ti
plt.savefig('H:/Thesis_Git/thesis_docs/Final_Thesis/Images/Complex-YOLO/pareto_optimality_slarti.pdf', bbox_inches='tight')

plt.show()
```
